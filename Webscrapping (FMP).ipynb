```python
# install modules and libraries
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
# download today news from FMP
# Use Apple stock here as an example
ticker = "AAPL"
url = "https://financialmodelingprep.com/financial-summary/" + ticker
request = requests.get(url)
print(request.text)
```

```python
parser = BeautifulSoup(request.text, "html.parser")
news_html = parser.find_all('a', {'class': "article"})
news_html[0]
```

```python
news_html[0].find('h4', {'class': "article__title-text"}).text
```

```python
# Create an empty list for later use
sentiments = []
# for loop here
for i in range(0, len(news_html)):
    # put the info back into the empty list
    sentiments.append(
            {
                'stock_name (ticker)': ticker,
                'date': news_html[i].find('h5', {'class': 'article__date'}).text,
                'title': news_html[i].find('h4', {'class': 'article__title-text'}).text,
                'extracted text': news_html[i].find('p', {'class': 'article__text'}).text
            }
        )
```

```python
sentiments[0]
```

```python
df = pd.DataFrame(sentiments)
df = df.set_index('date')
```
